```python
latest_year = df_awards.select(pl.col("yearID").max()).item()
train_awards = df_awards.filter(pl.col("yearID") < latest_year)
valid_awards = df_awards.filter(pl.col("yearID") == latest_year)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier # CART model
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score, make_scorer
```

```python
# These are the main offensive statistics:
X_cols = [ 'G_batting', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'SB', 'CS',  'BB', 'SO', 'IBB', 'HBP', 'SH', 'SF', 'GIDP'] 

TARGET_COL = 'Award_Category'

# Define X_train and y_train using the provided training data
X_train = train_awards[X_cols]
y_train = train_awards[TARGET_COL]

print(f"Features (X) shape: {X_train.shape}")
print(f"Target (y) shape: {y_train.shape}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
# Define the parameter grid
cart_param_grid = {
    'max_depth': [3, 5, 7, 9],
    'min_samples_leaf': [5, 10, 20],
    'criterion': ['gini', 'entropy']
}

# Initialize GridSearchCV
cart_grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=cart_param_grid,
    scoring='recall', # Optimize for the minority class (All-Star)
    cv=5, 
    verbose=1,
    n_jobs=-1
)

# Fit the grid search to the training data
print("Starting Grid Search for CART Model...")
cart_grid_search.fit(X_train, y_train)

# Print the best model details
print("\n Best CART Model")
print(f"Best Score (Recall): {cart_grid_search.best_score_:.4f}")
print(f"Best Parameters: {cart_grid_search.best_params_}")

best_cart_model = cart_grid_search.best_estimator_
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Define the parameter grid
rf_param_grid = {
    'n_estimators': [50, 100, 200], 
    'max_depth': [5, 10, None], 
    'min_samples_split': [2, 5],
    'class_weight': ['balanced'] 
}

# Initialize GridSearchCV
rf_grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=rf_param_grid,
    scoring='recall', # Optimize for the minority class
    cv=5,
    verbose=1,
    n_jobs=-1
)

# Fit the grid search to the training data
print("\nStarting Grid Search for Random Forest Model...")
rf_grid_search.fit(X_train, y_train)

# Print the best model details
print("\nBest Random Forest Model")
print(f"Best Score (Recall): {rf_grid_search.best_score_:.4f}")
print(f"Best Parameters: {rf_grid_search.best_params_}")

best_rf_model = rf_grid_search.best_estimator_
```

```python
from sklearn.metrics import recall_score, classification_report
import numpy as np
```

```python
X_cols = [
    'G_batting', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'SB', 'CS', 
    'BB', 'SO', 'IBB', 'HBP', 'SH', 'SF', 'GIDP'
] 
TARGET_COL = 'Award_Category'

X_valid = valid_awards[X_cols]
y_valid = valid_awards[TARGET_COL]

print(f"Validation Features (X) shape: {X_valid.shape}")
```

```python
models = {
    "CART Model": best_cart_model,
    "Random Forest": best_rf_model
}

print("\n Model Evaluation on Validation Data")
```

```python
for name, model in models.items():
    y_pred = model.predict(X_valid)
    
    labels = ["Not All-Star", "All-Star"]
    
    val_recall = recall_score(y_valid, y_pred, pos_label="All-Star")
```

```python
print(f"\nModel: **{name}**")
print(f"Validation Recall Score (All-Star): {val_recall:.4f}")
    
print("\nClassification Report:")
print(classification_report(y_valid, y_pred, target_names=labels))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.preprocessing import LabelBinarizer # Needed to handle string labels
```

```python
# LabelBinarizer converts the string labels ('All-Star', 'Not All-Star') 
# into the required 0/1 format for ROC calculation.
lb = LabelBinarizer()
y_valid_binary = lb.fit_transform(y_valid).ravel()
```

```python
# CART Model
cart_proba = best_cart_model.predict_proba(X_valid)[:, 1]

# Random Forest Model
rf_proba = best_rf_model.predict_proba(X_valid)[:, 1]
```

```python
# CART Metrics
cart_fpr, cart_tpr, _ = roc_curve(y_valid_binary, cart_proba)
cart_auc = roc_auc_score(y_valid_binary, cart_proba)
```

```python
# Random Forest Metrics
rf_fpr, rf_tpr, _ = roc_curve(y_valid_binary, rf_proba)
rf_auc = roc_auc_score(y_valid_binary, rf_proba)
```

```python
plt.figure(figsize=(10, 7))
# Plot CART ROC Curve
plt.plot(cart_fpr, cart_tpr, 
         label=f'CART (AUC = {cart_auc:.4f})', 
         color='blue', 
         linewidth=2)

# Plot Random Forest ROC Curve
plt.plot(rf_fpr, rf_tpr, 
         label=f'Random Forest (AUC = {rf_auc:.4f})', 
         color='green', 
         linewidth=2)

# Plot the baseline (random guessing)
plt.plot([0, 1], [0, 1], 
         linestyle='--', 
         color='red', 
         label='Chance (AUC = 0.50)', 
         alpha=0.6)

# Add plot labels and title
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR) / Recall')
plt.title('Comparative Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc='lower right')
plt.grid(linestyle=':', alpha=0.5)
plt.show()

print(f"\n Final AUC Scores on Validation Data ")
print(f"CART Model AUC: {cart_auc:.4f}")
print(f"Random Forest AUC: {rf_auc:.4f}")
```

Our goal was, essentially, to build a computer model that could look at a baseball player's raw hitting stats. Things like home runs, runs batted in (RBI), and batting average and then try to accurately predict if they would be chosen as an All-Star. What I did was test two different types of predictor tools:  CART and then Random Forest.All-Stars are pretty rare compared to the rest of the league, the most important job was maximizing Recall. Meaning making sure we identify and select the real All-Stars without missing any of them, then ran our final test on totally new player data, the Random Forest model actually clearly outperformed the simpler CART model. Proving to be significantly better at achieving our main goal of catching those all-star players and also showed a higher overall accuracy score (Area Under the Curve), which makes it the better method for predicting All-Star status based on batting performance alone.
